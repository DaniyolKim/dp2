```python
import pandas as pd
import numpy as npn
import datetime as dt
import plotly.express as px
import plotly.graph_objects as go
```

```python
df_org = pd.read_csv('https://raw.githubusercontent.com/DaniyolKim/dp2/main/result.csv')
df_org = pd.DataFrame({'time': pd.to_datetime(df_org['Unnamed: 0']), 'elec': df_org['elec']})
df_org['year'] = df_org['time'].dt.year
df_org['month'] = df_org['time'].dt.month
df_org['day'] = df_org['time'].dt.day
df_org['dayofweek'] = df_org['time'].dt.strftime('%A')
df_org['dayofweekIndex'] = df_org['time'].dt.dayofweek
df_org['hour'] = df_org['time'].dt.hour
df_org
```

```python
df_2021_04 = df_org[df_org['time'] >= dt.datetime(2021, 4, 1)]
df_2021_04
```

# **4월 전체 트렌드**

```python
fig1 = px.scatter(df_2021_04, x="time", y="elec", color='dayofweek')

fig2 = px.line(df_2021_04, x="time", y="elec")

fig3 = go.Figure(data=fig1.data + fig2.data)
fig3.show()
```

# 4월 전체 히스토그램

```python
fig = px.histogram(df_2021_04, x="elec", color='dayofweek')
fig.show()
```

```python
fig = px.histogram(df_2021_04, x="elec", color='hour')
fig.show()
```

```python
dayList = range(1, 31)
def showDailyEnergyPatterns(dataFrame):
  fig = go.Figure()
  for day in dayList:
    date = dataFrame['time'][dataFrame['day'] == day]
    name = ''
    try:
      name = str(day) + '(' + date.dt.strftime('%A').values[-1] + ')'
    except:
      name = ''
    
    fig.add_trace(go.Scatter(x=dataFrame['hour'][dataFrame['day'] == day],
                            y=dataFrame['elec'][dataFrame['day'] == day],
                            mode='lines',
                            name=name
                            ))
  fig.show()
```

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
def showDailyEnergyViolins(dataFrame):
  fig = go.Figure()
  for day in days:
    fig.add_trace(go.Violin(x=dataFrame['dayofweek'][dataFrame['dayofweek'] == day],
                            y=dataFrame['elec'][dataFrame['dayofweek'] == day],
                            name=day,
                            box_visible=True,
                            meanline_visible=True
                            ))
  fig.show()
```

# 패턴분석

## 4월 전체

```python
showDailyEnergyPatterns(df_2021_04)
showDailyEnergyViolins(df_2021_04)
```
