```python
import pandas as pd
import numpy as npn
import datetime as dt
import plotly.express as px
import plotly.graph_objects as go
```

```python
df_org = pd.read_csv('https://raw.githubusercontent.com/DaniyolKim/dp2/main/result.csv')
df_org = pd.DataFrame({'time': pd.to_datetime(df_org['Unnamed: 0']), 'elec': df_org['elec']})
df_org['year'] = df_org['time'].dt.year
df_org['month'] = df_org['time'].dt.month
df_org['day'] = df_org['time'].dt.day
df_org['dayofweek'] = df_org['time'].dt.strftime('%A')
df_org['dayofweekIndex'] = df_org['time'].dt.dayofweek
df_org['hour'] = df_org['time'].dt.hour
df_org
```

```python
df_2021_04 = df_org[df_org['time'] >= dt.datetime(2021, 4, 1)]
df_2021_04 = df_2021_04[:24*28]
df_2021_04
```

```python
dayList = range(1, 31)
def showDailyEnergyPatterns(dataFrame):
  fig = go.Figure()
  for day in dayList:
    date = dataFrame['time'][dataFrame['day'] == day]
    name = ''
    try:
      name = str(day) + '(' + date.dt.strftime('%A').values[-1] + ')'
    except:
      name = ''
    
    fig.add_trace(go.Scatter(x=dataFrame['hour'][dataFrame['day'] == day],
                            y=dataFrame['elec'][dataFrame['day'] == day],
                            mode='lines',
                            name=name
                            ))
  fig.show()
```

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
def showDailyEnergyViolins(dataFrame):
  fig = go.Figure()
  for day in days:
    fig.add_trace(go.Violin(x=dataFrame['dayofweek'][dataFrame['dayofweek'] == day],
                            y=dataFrame['elec'][dataFrame['dayofweek'] == day],
                            name=day,
                            box_visible=True,
                            meanline_visible=True
                            ))
  fig.show()
```

# 피드백 반영

## 4월 워킹데이 요일별 바이올린 차트

```python
#4월 워킹데이(월/화/수/목/금) 시간대별로 바이올린 차트
df_2021_04_wd = df_2021_04[df_2021_04['dayofweekIndex'] < 5]
showDailyEnergyViolins(df_2021_04_wd)
```

```python
#시간대별 중간값으로 0~24시까지의 사용량 바차트
def getProfileModelMed(dataFrame):
  return dataFrame.groupby(dataFrame['hour']).median()

def getProfileModelMean(dataFrame):
  return dataFrame.groupby(dataFrame['hour']).mean()

def plotBarChartPF(dataFrame):
  fig = go.Figure(data=[go.Bar(
    y=dataFrame['elec'], x=dataFrame.index
  )])
  fig.show()
```

## 4월 워킹데이 시간별 Median

```python
#4월 시간대별 median
df_2021_04_pf_med = getProfileModelMed(df_2021_04_wd);
plotBarChartPF(df_2021_04_pf_med)
#4월 시간대별 mean
# df_2021_04_pf_mean = getProfileModelMean(df_2021_04_wd);
# plotBarChart(df_2021_04_pf_mean)
```

```python
def plotBarChartElec(pfModel, targetDF, year, month, date):
  fig = go.Figure()
  fig.add_trace(go.Bar(
    x=pfModel.index,
    y=pfModel['elec'],
    name='Profile',
    marker_color='indianred'
  ))

  dateName = dt.datetime(year, month, date)
  tdf = targetDF[(targetDF['time'] >= dt.datetime(year, month, date, 0)) & (targetDF['time'] <= dt.datetime(year, month, date, 23))]
  fig.add_trace(go.Bar(
      x=tdf['hour'],
      y=tdf['elec'],
      name=dateName.strftime('%Y-%m-%d'),
      marker_color='lightsalmon'
  ))
  
  fig.update_layout(barmode='group', xaxis_tickangle=-45)
  fig.show()
```

## Median profile과 4월 5일 비교

```python
plotBarChartElec(df_2021_04_pf_med, df_2021_04_wd, 2021, 4, 5)
```

## Median profile과 4월 9일 비교(창립 기념일)

```python
plotBarChartElec(df_2021_04_pf_med, df_2021_04_wd, 2021, 4, 9)
```
